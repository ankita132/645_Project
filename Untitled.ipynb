```python
viewCombinations = set()
i=0
for a in attributes:
    for m in measures:
        if a != m:
            for agg in aggregations:
                viewCombinations.add((a,m,agg))

print(len(viewCombinations))
```

```python
# calculation of utility based on KL-divergence
def calcUtility(p, q):
    # assuming dictionaries
    assert len(p) == len(q)
    utility = 0
    for k in p.keys():
        v1 = float(p[k])
        v2 = float(q[k])
        utility += (v1 * math.log(v1/v2))
    
    return utility
```

```python
# create the probability distribution of the target and reference query results
def getP(respr, respt):
    target_set = {}
    reference_set = {}
    for v in respt:
        target_set[v[0]] = v[1]
    for v in respr:
        reference_set[v[0]] = v[1]
    
    sumT = 0
    sumR = 0
    for k in target_set.keys():
        if k not in reference_set.keys() or target_set[k] == 0 or reference_set[k] == 0:
            target_set[k] = 0
        else:
            sumT += target_set[k]
        
    for k in reference_set.keys():
        if k not in target_set.keys() or target_set[k] == 0 or reference_set[k] == 0:
            reference_set[k] = 0
        else:
            sumR += reference_set[k]
        
    
    userData = {}
    refData = {}
    for k in target_set.keys():
        if target_set[k] != 0:
            userData[k] = target_set[k]/sumT
    
    
    for k in reference_set.keys():
        if reference_set[k] != 0:
            refData[k] = reference_set[k]/sumR
    
    return userData, refData
```

```python
%%time
# Top 5 utility using brute force, getting utility for all the views
allUtilities = []
N = 5

for v in viewCombinations:
    a,m,f = v[0],v[1],v[2]
    sqlt = f"select {a}, {f}({m}) from Adult where {createUserTable.getWhereCondition()} group by {a}"
    sqlr = f"select {a}, {f}({m}) from Adult where {createReferenceTable.getWhereCondition()} group by {a}"
    cursor.execute(sqlr)
    respr = cursor.fetchall()
    cursor.execute(sqlt)
    respt = cursor.fetchall()
    userData, refData = getP(respr, respt)
    utility = calcUtility(userData, refData)
    allUtilities.append(((a,m,f), utility))

ordered = sorted(allUtilities,key=lambda x: x[1], reverse=True)
print("Top 5 high utility views with utility score by brute force")
print(ordered[0:5])
print("Average of top 5 utilities : ", sum([pair[1] for pair in ordered[0:5]])/5)
# for view in ordered[0:5]:
#     plotView(view[0])
   
# [(('occupation', 'capital_gain', 'avg'), 0.7164254431169454), 
# (('education_num', 'capital_gain', 'avg'), 0.6805703651309709), 
# (('education', 'capital_gain', 'avg'), 0.6805703651309709), 
# (('education_num', 'capital_gain', 'sum'), 0.6199569423443795), 
# (('education', 'capital_gain', 'sum'), 0.6199569423443795)]
```

```python
# Combine multiple aggregates list
def getCombinedMultipleAggregates(): 
    multipleAggregates = {}
    for comb in viewCombinations:
        a = comb[0]
        if a not in multipleAggregates:
            multipleAggregates[a] = [comb]
        else:
            multipleAggregates[a].append(comb)
    return multipleAggregates
```

```python
# get all number of records in Adult
cursor.execute('select count(*) from Adult')
numRecords = cursor.fetchall()[0][0]
```

```python
%%time
##################### Shared optimization WITH Combine Multiple Aggregate and Rewriting queries and perform pruning
cumKl={}
combinedAggregates = getCombinedMultipleAggregates()
splitCount, i = 20, 0
phaseData = int(numRecords/splitCount)
delta = 0.001
N = splitCount
M = 0
k = 5
prune_views = []
to_prune_combinations = viewCombinations
while i <= numRecords:
    view_stats = {}
    M = M+1
    for a in combinedAggregates.keys():
        aggregates = []
        numAggregates = len(combinedAggregates[a])
        for comb in combinedAggregates[a]:
            m=comb[1]
            f=comb[2]
            aggregates.append(f'{f}({m})')
        selectStr = ','.join(aggregates)
#         print(selectStr)
        sql = f"select {a}, {selectStr}, case when {createUserTable.getWhereCondition()} then 1 else 0 end as g1 from (select * from Adult offset {i} limit {phaseData}) group by {a}, g1 order by {a}, g1"
        cursor.execute(sql)
        res = cursor.fetchall()
#         print(sql)
        respt, respr = [[] for _ in range(numAggregates)], [[] for _ in range(numAggregates)]
#         print(len(respt), len(respr))
        for ele in res:
            for j in range(numAggregates):
                if ele[-1] == 1:
                    respt[j].append((ele[0], ele[j+1]))
                else:
                    respr[j].append((ele[0], ele[j+1]))

        for j in range(numAggregates):
            userData, refData = getP(respr[j], respt[j])
            # Calculate utility
            utility = calcUtility(userData, refData)
            curComb = combinedAggregates[a][j]
            m = curComb[1]
            f = curComb[2]
            view_stats[a,m,f] = defaultdict(dict)
            if (a,m,f) in cumKl:
                cumKl[(a,m,f)] = cumKl[(a,m,f)] * (1-1/(M)) + utility * (1/M)
            else:
                cumKl[(a,m,f)] = utility
            # Calculate confidence interval
            if(M>1): 
                numr = (1-(M-1)/N)*(2*log(log(M)))+log(pi**2/(3*delta))
                denr = 2*M
                epsilon = (numr/denr)**0.5
                view_stats[a,m,f][0] = cumKl[(a,m,f)] ## mean
                view_stats[a,m,f][1] = view_stats[a,m,f][0] - epsilon ## lower bound
                view_stats[a,m,f][2] = view_stats[a,m,f][0] + epsilon ## upper bound
#                 print(view_stats[a,m,f], epsilon)
    # Prune  
    if M>1:
        view_stats_sorted = dict(sorted(view_stats.items(), key=lambda e: e[1][2], reverse=True))
        ub_topk = dict(list(view_stats_sorted.items())[0: k])
        lb_min = min([ub_topk[x][1] for x in ub_topk.keys()])
        not_topk = dict(list(view_stats_sorted.items())[k:])
        for v in not_topk.keys():
            if(not_topk[v][2]<lb_min):
                prune_views.append(v)
            
        to_prune_combinations = list(set(to_prune_combinations).difference(prune_views))
        print(f"{len(prune_views)} pruned in phase {M}, remaining views: {len(to_prune_combinations)}")
                  
    i += phaseData + 1

utility_vs_view = []
for view in to_prune_combinations:
    utility_vs_view.append((view, cumKl[view]))

ordered = sorted(utility_vs_view,key=lambda x: x[1], reverse=True)
print("Top 5 high utility by Shared optimization WITH Combine Multiple Aggregate and pruning")
print(ordered[:5])
print("Average of top 5 utilities : ", sum([pair[1] for pair in ordered[0:5]])/5)
# for view in ordered[0:5]:
#     plotView(view[0])
```
